## Preprocessing

````python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
````

````python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])

application_df.head()
````

````python
# Determine the number of unique values in each column.
application_df.nunique()
````

````python
# Look at APPLICATION_TYPE value counts for binning
app_type_counts = application_df['APPLICATION_TYPE'].value_counts()

app_type_counts
````

````python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = [atype for atype, num in app_type_counts.items() if num < 500]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
````

````python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()

classification_counts
````

````python
# You may find it helpful to look at CLASSIFICATION value counts >1
[[label, val] for label, val in application_df['CLASSIFICATION'].value_counts().items() if val > 1 ]
````

````python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = [atype for atype, num in classification_counts.items() if num < 1000]

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
````

````python
application_df['INCOME_AMT'].value_counts()
````

````python
# Convert categorical data to numeric with `pd.get_dummies`
categorical_cols = ['APPLICATION_TYPE',
                    'AFFILIATION',
                    'CLASSIFICATION',
                    'USE_CASE',
                    'ORGANIZATION',
                    'INCOME_AMT',
                    'SPECIAL_CONSIDERATIONS'
                    ]
encoded_cats = pd.get_dummies(application_df[categorical_cols].drop(columns=['SPECIAL_CONSIDERATIONS']))

encoded_cats.head(10)
````

````python
application_df[['AFFILIATION', 'IS_SUCCESSFUL']].value_counts()
````

````python
application_df[['USE_CASE', 'IS_SUCCESSFUL']].value_counts()
````

````python
merged_df = pd.concat([application_df, encoded_cats], axis=1).drop(columns=categorical_cols)

print(len(merged_df))
merged_df.head()
````

````python
merged_df.columns
````

````python
merged_df[['STATUS', 'IS_SUCCESSFUL']].value_counts()
````

````python
# merged_df[['SPECIAL_CONSIDERATIONS_Y', 'IS_SUCCESSFUL']].value_counts()
````

````python
# Split our preprocessed data into our features and target arrays
X = merged_df.drop(columns=['IS_SUCCESSFUL', 'STATUS'])
y = merged_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
````

````python
X.columns
````

````python
X
````

````python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
````

## Compile, Train and Evaluate the Model

````python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential([
    tf.keras.layers.Dense(80, input_dim=X.shape[1]),
    tf.keras.layers.Dense(30, activation='tanh'),
    tf.keras.layers.Dense(20, activation='sigmoid'),
    tf.keras.layers.Dense(30, activation='linear'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

nn.build()

# First hidden layer
#  YOUR CODE GOES HERE

# Second hidden layer
#  YOUR CODE GOES HERE

# Output layer
#  YOUR CODE GOES HERE

# Check the structure of the model
nn.summary()
````

````python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
nn.fit(X_train_scaled, y_train, epochs=15, validation_data=(X_test_scaled,y_test))
````

````python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
````
